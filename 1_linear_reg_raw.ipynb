        Making a very basic linear regression model

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from numpy import *
```

```python
x = [1, 3, 5, 6, 8, 12, 20, 11, 12]
y = [4, 3, 7, 7, 11, 14, 22, 11, 10]
x = np.array(x)
y = np.array(y)
```

```python
plt.scatter(x, y)
```

```python
m, c = np.random.random(), np.random.random()
```

Normalize x and y arrays

```python
x = x/10
y = y/10
alpha = 0.1
```

```python
for k in range(1000):
    y_predict = m*x + c
    
    #gradient descent for theta0
    error =  y_predict - y
    error_avg = np.sum(error)/len(error)
    
    #gradient descent for theta1
    error_mod = error*x
    error_mod_sum = np.sum(error_mod)/len(error)
    
    c = c - alpha*error_avg
    m = m - alpha*error_mod_sum
    
    cost = np.sum((y_predict - y)**2)/2*len(y)
    print(cost)
    
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict)
```

```python
query = input("Enter the value of x for prediction : ")
query = (float(query)) 
```

```python
y_calc = m*query + c
```
